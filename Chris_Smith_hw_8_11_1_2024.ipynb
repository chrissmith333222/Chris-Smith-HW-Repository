#### Find the signals and create a new column that categorizes the incidents.

There is no "right" answer here, just a way for you as a emerging data scientist to think about how to parse data and categorize it. Just declare categories and justify your logic! I am as interested in your reasoning, as I am in your code!

```python
#Yout code here:
 #ChatGPT
#Your code here
#Importing pandas
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file with specified encoding
csv_file_path = 'RAND_Database_of_Worldwide_Terrorism_Incidents.2.csv'
df = pd.read_csv(csv_file_path, encoding='ISO-8859-1')

# Define keywords for each type of terrorism
political_keywords = ['political', 'government', 'policy', 'state', 'election']
religious_keywords = ['religious', 'faith', 'sectarian', 'holy', 'jihad', 'islam', 'christian']
economic_keywords = ['economic', 'finance', 'money', 'business', 'trade', 'corporate']

# Function to classify terrorism type based on keywords in description
def classify_terrorism(description):
    # Check if the description is NaN (missing)
    if pd.isna(description):
        return 'Other Terrorism'  # Categorize NaN values as "Other Terrorism"
    
    description = description.lower()  # Convert to lowercase for uniformity
    if any(keyword in description for keyword in political_keywords):
        return 'Political Terrorism'
    elif any(keyword in description for keyword in religious_keywords):
        return 'Religious Terrorism'
    elif any(keyword in description for keyword in economic_keywords):
        return 'Economic Terrorism'
    else:
        return 'Other Terrorism'

# Apply the classification function to the 'Description' column
df['Terrorism_Type'] = df['Description'].apply(classify_terrorism)  # Adjusted to 'Description' based on your column name

# Count occurrences of each terrorism type
terrorism_counts = df['Terrorism_Type'].value_counts()

# Plotting the results
plt.figure(figsize=(10, 6))
terrorism_counts.plot(kind='bar', color=['blue', 'green', 'orange', 'red'])

# Adding labels and title
plt.xlabel("Type of Terrorism")
plt.ylabel("Number of Incidents")
plt.title("Classification of Terrorism Incidents by Type")

# Show the plot
plt.show()
```
