### Read vs Readlines
Above, we see that we can grab all the content of the fie using read(). This is a great way to go if we want to work with the file content as a whole, for example, searching for a word in the entire file, or processing word frequency distribution on a presidential speech.

But what if the the file is very large? Teading the entire content into memory can be inefficient or even cause memory issues. 

Or what if we want to process the file line by line and pull out spaces, or other extraneous info that we don't want? 

Readlines to the rescue! 
Readlines reads the file line by line and returns a list of strings, where each string represents a line from the file.

Use readlines() when you need to work with the file line by line, and you want to process or iterate over the lines.

Here is an example:

```python
with open('first_lines_leader.txt', 'r') as file:
    lines = file.readlines()  # Reads the file and returns a list of lines
for line in lines:
    print(line.strip())  # Strip removes newline characters
```

### Alternative for very large files: 
I want to introduce you to this idea, because ChatGPT might suggest it, so I want you to be exposed and know what's going on.
For very large files, neither read() nor readlines() may be efficient. Instead, you can read the file line by line without loading the whole file into memory using what is called a file object iterator:



```python

with open('first_lines_leader.txt', 'r') as file:
    for line in file:
        print(line.strip())
```

That said, you should use read() or readlines() in the majority of your cases as your file size will be moderately large at most.

### Some summary:

Opening a File:

Modes:
'r' = read
'w' = write (overwrites the file if it exists)
'a' = append (adds to the end of the file)
'b' = binary mode (for non-text files)

```python
#chatGPT
#import of NLTK inaugural works in the same manner as last lesson with generation of a FreqDist plot

import nltk
from nltk.corpus import inaugural
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Download the required NLTK datasets (Inaugural speeches, stopwords, tokenizer)
nltk.download('inaugural')
nltk.download('stopwords')
nltk.download('punkt')

# Set of common English stopwords (like "the", "and", etc.)
stopwords = set(stopwords.words('english'))

# Function to clean speech text by removing stopwords and non-alphabetic characters
# and return the frequency distribution of words
def clean_text(text, stopwords):
    clean_text = [word.lower() for word in text if word.isalpha() and word.lower() not in stopwords]
    return FreqDist(clean_text)

# Function to process each speech: clean the text, save the word frequency to a file,
# and create/save a plot of the 20 most common words.
def process_speech(speech_name, filename_base):
    # Load the words from the speech file
    speech_words = inaugural.words(speech_name)
    
    # Clean the words and get frequency distribution
    fdist = clean_text(speech_words, stopwords)
    
    # Save the frequency distribution to a text file
    save_fdist_to_file(fdist, f'{filename_base}_fdist.txt')
    
    # Plot and save the 20 most common words as a PNG image
    plt.figure()
    fdist.plot(20, title=f"{filename_base.replace('_', ' ').title()} Word Use")  # Title the plot
    plt.savefig(f'{filename_base}_plot.png')  # Save the plot as an image file

# Function to save the 20 most common words and their frequencies into a text file
def save_fdist_to_file(fdist, filename):
    with open(filename, 'w') as f:
        for word, frequency in fdist.most_common(20):  # Get the top 20 most frequent words
            f.write(f'{word}: {frequency}\n')  # Write each word and its frequency to the file

# List of inaugural speeches we want to process along with the desired filenames
speeches = [
    ("1789-Washington.txt", "washington_inaugural"),
    ("1861-Lincoln.txt", "lincoln_first_inaugural"),
    ("1865-Lincoln.txt", "lincoln_second_inaugural")
]

# Loop through each speech in the list and process it
for speech_file, filename_base in speeches:
    process_speech(speech_file, filename_base)
    
print("Synopsis: We can analyze these speeched in terms of Presidential commentary in a period of 1. Prospertity and new growth, 2. Crisis and growing division, and 3. Conflict and Civil War.  Washington's speech refelects optimistim and speaks of 'government' and 'citizens', an illusion to the new creation of the American state.  Lincoln's first inaugural emphasizes 'union', 'the people', and the individual 'states' while his second inaugural emphasizes 'war' and invokes 'God' due to the dire situation in the country")
```

```python
#chatGPT
#import beautiful soap to import text from the web

import requests
from bs4 import BeautifulSoup

# Step 1: Set up headers for the request to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

# Step 2: Fetch the webpage content using the 'requests' library with a user-agent header
base_url = "http://www.en.kremlin.ru/supplement/5770"
try:
    r = requests.get(base_url, headers=headers)
    r.raise_for_status()  # Check if the request was successful
except requests.exceptions.RequestException as e:
    print(f"Error fetching content from {base_url}: {e}")
else:
    # Step 3: Parse the HTML content using BeautifulSoup with the 'html.parser'
    soup = BeautifulSoup(r.text, 'html.parser')

    # Step 4: Find all paragraph elements ('<p>') in the HTML
    paragraphs = soup.find_all('p')

    if paragraphs:
        # Step 5: Create a list called 'document' to store paragraph text
        document = []
        
        # Step 6: Iterate over the paragraphs and extract text from each
        for p in paragraphs:
            text = p.get_text()  # Get the text content of the paragraph
            document.append(text)  # Append the text to the document list

        # Display the content that has been added to the document list
        for i, para in enumerate(document):
            print(f"Paragraph {i+1}: {para}\n")

        print("Content successfully written to the 'document' list.")
    else:
        print("No paragraphs found on the page.")
```

### Intro to Summarization
Summarization has been a thorny problem for data scientists over the past decade. There have been many approaches, but most, until ChatGPT and it's use of the transformers' architecture, they have been unsatisfactory. There are two main types of summarization:

Extractive Summarization: Finding the most important sentences and rendering them verbatim.

Abstractive Summarization: Taking the most important ideas from a document and generating a summary based on them

There have been many different algorithms that have attempted summarization. Let's learn about one of the most used and deploy it in python.

#### Luhns Heuristic Method
This is one of the earliest approaches of text summarization. This is related to our manual interpretation of the importance of word frequency in a document.Basically, just as we did, Luhn recognized that word frequency and word significance is highly correlated, but only "important" words. The assumption that the most important sentences are those with the most significant words. Here is a nice short article that explains the algo in an accessible way: https://blog.fastforwardlabs.com/2016/03/25/h.p.-luhn-and-the-heuristic-value-of-simplicity.html
